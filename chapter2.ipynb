## Binomial Distribution

```python
from scipy.stats import binom
from math import exp
import matplotlib.pyplot as plt
import numpy as np
n = 1000 # number of trials
p = 0.45 # probability of success
probabilities = [] # list of probabilities
for k in range(n+1):
    probability = binom.pmf(k, n, p)
    probabilities.append(probability)
    # print('{0}-{1}'.format(k, probability))

x = np.arange(0, 1001) 
y = np.array(probabilities)

plt.plot(x, y, linestyle='-')
plt.show()
```

## Beta Distribution

Given $a$ successes and $b$ failures, we can calculate the probability that the probability of success is below some threshold $p$ using the beta distribution. That is, if we have 7 successes and 3 failures, what's the chance the underlying probability of success of the operation was, in fact, 70%?

```python
from scipy.stats import beta
import matplotlib.pyplot as plt
import numpy as np

a = 7
b = 3

# p = beta.cdf(.70, a, b)
# print(p)
probabilities=[]
for k in range(0, 100, 5):
    probability =  beta.pdf(k/100, a, b)
    probabilities.append(probability)
    #print('{0}-{1}'.format(k, probability))

x = np.arange(0, 1, .05) 
y = np.array(probabilities)

# plot likelihood of a given probability of that number of successes and failures
plt.plot(x, y, linestyle='-')
plt.show()
```
